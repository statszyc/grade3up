```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("iris(1).csv")
```

```python
with sns.axes_style("darkgrid"):
    ax=sns.pairplot(df[["sepal_length","sepal_width","petal_length","petal_width","species"]],hue="species")
    ax=ax.map_lower(sns.regplot)
    ax=ax.map_upper(sns.regplot)
    ax=ax.map_diag(sns.kdeplot)
```

第三题

```python
df2=pd.read_csv("gapminder(1).csv")
df2=df2.iloc[:,1:]
```

3.1

```python
sns.lmplot(x="gdpPercap",y="lifeExp",data=df2,logx=True)
plt.title("gdpPercap~lifeExp with all data")
```

```python
sns.lmplot(x="gdpPercap",y="lifeExp",data=df2[df2["continent"]=="Asia"],logx=True)
plt.title("gdpPercap~lifeExp with data of Asia")
```

```python
sns.lmplot(x="gdpPercap",y="lifeExp",data=df2[df2["country"]=="China"],logx=True)
plt.title("gdpPercap~lifeExp with data of China")
```

3.2

```python
import numpy as np
sns.lmplot(x="year",y="lifeExp",data=df2,x_estimator=np.mean,hue="continent")
```

3.3

```python
ax=sns.histplot(x="lifeExp",
                data=df2,
                kde=True,
                stat="density",
                element="step",
                kde_kws=dict(cut=3),
                alpha=.4,
                edgecolor=(1, 1, 1, .4),
                hue="continent")
```
