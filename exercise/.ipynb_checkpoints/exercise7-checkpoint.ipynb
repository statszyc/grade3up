1. 给定dataframe  df=pd.DataFrame({"vals": np.random.RandomState(31).randint(-30, 30, size=15), "grps": np.random.RandomState(31).choice(["A", "B"], 15)})
给df新增一列patched_vals, 如果vals相应行非负，则patched_vals等于vals，否则等于组平均值(按照grps分组)。


```python
import numpy as np
import pandas as pd
df=pd.DataFrame({"vals": np.random.RandomState(31).randint(-30, 30, size=15), "grps": np.random.RandomState(31).choice(["A", "B"], 15)})
df
```

```python
from dataclasses import replace
from textwrap import fill


means=df.groupby("grps").mean()
abmean=df["grps"].map(lambda x: means.iloc[0,0] if x=="A" else means.iloc[1,0])
df["patched_vals"]=df["vals"].map(lambda x: x if x>=0 else np.nan)
# df["patched_vals"]+=abmean*df["patched_vals"].isna()
df["patched_vals"]=df["patched_vals"].add(abmean*df["patched_vals"].isna(),fill_value=0)
df
```

2. 读入iris.csv文件到dataframe iris并增加两列：1. petal.ratio=petal.length/petal.width, 2.  sepal.ratio=sepal.length/sepal.width. 使用groupby计算不同品种鸢尾花(variety)
的petal.ratio和sepal.ratio的mean, median, minimum, maximum, standard deviation。


```python
df2=pd.read_csv("iris(1).csv")
df2["petal.ratio"]=df2["petal.length"]/df2["petal.width"]
df2["sepal.ratio"]=df2["sepal.length"]/df2["sepal.width"]
df2
```

```python
df2[["petal.ratio","sepal.ratio"]].groupby(df2["variety"]).describe()
```

3. 对于2题中的dataframe iris，随机设置sepal.length,sepal.width,petal.length,petal.width这四列中20个值为nan，然后使用组均值来填充这20个值，并更新petal.ratio和sepal.ratio。

```python
df2.loc[np.random.choice(df2.shape[0],20,replace=False),"sepal.length"]=np.nan
df2.loc[np.random.choice(df2.shape[0],20,replace=False),"sepal.width"]=np.nan
df2.loc[np.random.choice(df2.shape[0],20,replace=False),"petal.length"]=np.nan
df2.loc[np.random.choice(df2.shape[0],20,replace=False),"petal.width"]=np.nan
```

```python
df2["sepal.length"]=df2["sepal.length"].fillna(df2["sepal.length"].mean())
df2["sepal.width"]=df2["sepal.width"].fillna(df2["sepal.width"].mean())
df2["petal.length"]=df2["petal.length"].fillna(df2["petal.length"].mean())
df2["petal.width"]=df2["petal.width"].fillna(df2["petal.width"].mean())
```
