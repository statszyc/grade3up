```r
library(ggplot2)
```

```r
ggplot(data = iris,aes(x=Petal.Length,fill=Species))+
geom_density(alpha=0.5)
```

```r
ggplot(data = iris,aes(x=Petal.Length,fill=Species))+
geom_histogram()
```

```r
ggplot(data = iris,aes(x=Species,y=Petal.Length))+
geom_boxplot()+
geom_point(position="jitter")
```

```r
ggplot(data = iris,aes(x=Sepal.Length,y=Sepal.Width,color=Species,shape=Species))+
geom_point(size=3)
```

```r
df1=data.frame(value=c(apply(iris[iris$Species=='setosa',][,1:4],2,sum),
apply(iris[iris$Species=='versicolor',][,1:4],2,sum),
apply(iris[iris$Species=='virginica',][,1:4],2,sum)),
variable=rep(c('Sepal.Length','Sepal.Width','Petal.Length','Petal.Width'),3),
Species=rep(c("setosa","versicolor","virginica"),each=4))
```

```r
ggplot(data=df1,aes(x=Species,y=value,fill=variable))+
geom_histogram(position = "stack",stat = 'identity')
```

```r
df2=data.frame(value=c(apply(iris[iris$Species=='setosa',][,1:4],2,mean),
apply(iris[iris$Species=='versicolor',][,1:4],2,mean),
apply(iris[iris$Species=='virginica',][,1:4],2,mean)),
variable=rep(c('Sepal.Length','Sepal.Width','Petal.Length','Petal.Width'),3),
Species=rep(c("setosa","versicolor","virginica"),each=4))
```

```r
ggplot(data=df2,aes(x=Species,y=value,fill=variable))+
geom_histogram(position = "dodge",stat = 'identity')
```

```r
ggplot(data=iris,aes(x=Sepal.Length,y=Sepal.Width,color=Species))+
geom_point()+
facet_grid(Species~.)
```
