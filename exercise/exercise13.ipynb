# 第一问

```r
quadsol <- function(k1, k2, k3) {
    if (missing(k1) | missing(k2) | missing(k3)) {
        return("You missed at least 1 parameter")
    }
    if (k2^2 - 4 * k1 * k3 < 0) {
        print("No real roots could be found, here are two complex roots: ")
        return(c((-k2 - sqrt(-(k2^2 - 4 * k1 * k3)) * 1i) / (2 * k1), (-k2 + sqrt(-(k2^2 - 4 * k1 * k3)) * 1i) / (2 * k1)))
    }
    if (k2^2 - 4 * k1 * k3 == 0) {
        return(-k2 / (2 * k1))
    }
    if (k2^2 - 4 * k1 * k3 > 0) {
        return(c((-k2 - sqrt(k2^2 - 4 * k1 * k3)) / (2 * k1), (-k2 + sqrt(k2^2 - 4 * k1 * k3)) / (2 * k1)))
    }
}
```

```r
quadsol(-1, -2, 0)
```

```r
quadsol(1, 2, 1)
```

```r
quadsol(4, 2, 3)
```

```r
quadsol(-1, -2)
```

## c

## d

```r
simout <- rpois(n = 260, lambda = 107)
hist(simout, xlim = c(60, 150))
```

下面用两种方法比较模拟的分布和理论的泊松分布

```r
hist(simout, freq = FALSE, xlim = c(60, 150))
points(seq(60, 150, by = 1), dpois(seq(60, 150, by = 1), lambda = 107), col = 2, lty = 2, lwd = 2)
```

可见两者形状基本吻合

```r
qqplot(simout, qpois(ppoints(simout), lambda = 107))
abline(0, 1, col = 2, lty = 2)
```

大部分散点落在直线上或直线周围，说明两个分布基本一样

# 第三问

## i

```r
mymean <- 10
mysd <- sqrt(2)
pnorm(11, mean = mymean, sd = mysd) - pnorm(9.5, mean = 10, sd = sqrt(2))
```

$P(9.5\leq X\leq 11)\approx 0.398413133990642$

## ii

```r
mystd1 <- (9.5 - mymean) / mysd
mystd2 <- (11 - mymean) / mysd
#标准化
pnorm(mystd2) - pnorm(mystd1)
```

可见与i中的概率相同

## iii

```r
qnorm(0.025, mean = mymean, sd = mysd)
```
