```python
import seaborn as sns
def plot_heatmap(x,y,x_opts,y_opts):
    column1=x_opts
    index1=y_opts
    df0=df_using[[x,y]]
    dff=pd.DataFrame(np.zeros((len(index1),len(column1))),index=pd.Index(index1,name=x),columns=pd.Index(column1,name=y))
    for sample in range(df_using.shape[0]):
        for i in index1:
            for j in column1:
                if bool(df_using[sample:sample+1][x].values==j):
                    if bool(df_using[sample:sample+1][x].values==j):
                        if df_using[sample:sample+1][y].values==i:  #频数统计
                            dff.loc[i,j]+=1
                
    result=pd.DataFrame(dff.values.T,columns=index1,index=column1)

    sns.heatmap(result, annot=True, fmt="g", cmap='OrRd')
    plt.yticks(rotation=0)
    plt.show()
```
