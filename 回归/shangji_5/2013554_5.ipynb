## 源代码：

```python
import pandas as pd
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
df=pd.read_csv("6-9.txt",sep='\s+',names=['potency','temp','value'])
df
```

```python
model=smf.ols("value~C(potency)+C(temp)",data=df).fit()
anova_table=anova_lm(model,type=2)
pd.DataFrame(anova_table)
```

## 统计分析：

方差分析表如上。浓度的p值为0.013761\<0.05，说明在显著性水平$\alpha$=0.05下浓度对产品得率的影响有显著差异。

温度的p值为0.201340 \> 0.05，说明在显著性水平$\alpha$=0.05下温度对产品得率的影响无显著差异。
